```python
from torchvision.datasets import CIFAR10
import matplotlib.pyplot as plt
import numpy as np
```

```python
root="data/"
```

```python
train_data = CIFAR10(root, download=True)

x = train_data.data
y = np.array(train_data.targets)

print(x.shape)
print(y.shape)
```

```python
image = x[4]

# plt.axis("off")
fig, ax = plt.subplots(figsize=(2, 2))
ax.axis("off")
ax.imshow(image)
plt.show()
```

```python
# Flip color channels.

image = x[4][...,::-1]

# plt.axis("off")
fig, ax = plt.subplots(figsize=(2, 2))
ax.axis("off")
ax.imshow(image)
plt.show()
```

```python
# Flip left right.

image = x[4][:, ::-1, :]

# plt.axis("off")
fig, ax = plt.subplots(figsize=(2, 2))
ax.axis("off")
ax.imshow(image)
plt.show()
```

```python
n = len(x)
flips = np.random.binomial(1, 0.5, size=(n))
flipped_images = np.array([x[i][:, ::-1, :] if flips[i] else x[i] for i in range(n)])
```

```python
flipped_images.shape
```

```python
for i in range(5):
    print(bool(flips[i]))
    image1 = x[i]
    image2 = flipped_images[i]
    # plt.axis("off")
    fig, ax = plt.subplots(1, 2, figsize=(2, 4))
    ax[0].axis("off")
    ax[1].axis("off")
    ax[0].imshow(image1)
    ax[1].imshow(image2)
    plt.show()
```
